### Global Superstore Regional Audit

This section generates a simulated dataset for the audit and then creates a 2x2 dashboard to answer key architectural questions regarding efficiency, risk, and category health across major territories.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(101)

# Generate Audit Data
n_rows = 500
regions = ['East', 'West', 'Central', 'South']
categories = ['Technology', 'Furniture', 'Office Supplies']

data = {
    'Region': np.random.choice(regions, n_rows),
    'Category': np.random.choice(categories, n_rows),
    'Sales': np.random.lognormal(mean=5, sigma=1, size=n_rows) * 10,
    'Profit': np.random.normal(loc=100, scale=400, size=n_rows),
    'Discount': np.random.choice([0, 0.1, 0.2, 0.5, 0.8], n_rows, p=[0.4, 0.3, 0.15, 0.1, 0.05])
}

df = pd.DataFrame(data)

print("Simulated dataset created successfully. Displaying the first 5 rows:")
display(df.head())
```

```python
# --- START ARCHITECTURAL CODE HERE ---
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Super-Title
fig.suptitle('Superstore Regional Audit: Q4 Performance', fontsize=20)

# Sub-Question 1: The Discount Trap (Top-Left)
# Histogram of Discount with Mean Discount line
axes[0, 0].hist(df['Discount'], bins=5, edgecolor='black', alpha=0.7)
mean_discount = df['Discount'].mean()
axes[0, 0].axvline(mean_discount, color='red', linestyle='dashed', linewidth=2, label=f'Mean Discount: {mean_discount:.2%}')
axes[0, 0].set_title('Distribution of Discounts')
axes[0, 0].set_xlabel('Discount (%)')
axes[0, 0].set_ylabel('Number of Orders')
axes[0, 0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))
axes[0, 0].legend()


# Sub-Question 2: Regional Profitability Benchmarks (Top-Right)
# Bar Chart of Total Profit for each Region
regional_profit = df.groupby('Region')['Profit'].sum().reindex(regions) # Ensure consistent order

region_colors = {'East': 'blue', 'West': 'green', 'Central': 'orange', 'South': 'purple'}
bars = axes[0, 1].bar(regional_profit.index, regional_profit.values, color=[region_colors[r] for r in regional_profit.index])
axes[0, 1].set_title('Total Profit by Region')
axes[0, 1].set_xlabel('Region')
axes[0, 1].set_ylabel('Total Profit ($)')
axes[0, 1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))

# Annotate bars with profit values
for bar in bars:
    yval = bar.get_height()
    axes[0, 1].annotate(f'${yval:,.0f}',
                xy=(bar.get_x() + bar.get_width() / 2, yval),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')


# Sub-Question 3: Sales Efficiency (Bottom-Left)
# Scatter Plot of Sales vs. Profit, colored by Category
for category in categories:
    subset = df[df['Category'] == category]
    axes[1, 0].scatter(subset['Sales'], subset['Profit'], label=category, alpha=0.6)
axes[1, 0].axhline(y=0, color='red', linestyle='--', linewidth=1, label='Break-Even Point (Profit = 0)')
axes[1, 0].set_title('Sales vs. Profit by Category')
axes[1, 0].set_xlabel('Sales ($)')
axes[1, 0].set_ylabel('Profit ($)')
axes[1, 0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))
axes[1, 0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))
axes[1, 0].legend(title='Category')


# Sub-Question 4: Category Volatility (Bottom-Right)
# Boxplot of Profit for each Category
boxplot_data = [df[df['Category'] == cat]['Profit'] for cat in categories]
boxes = axes[1, 1].boxplot(boxplot_data, labels=categories, patch_artist=True)
axes[1, 1].set_title('Profit Distribution by Category')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Profit ($)')
axes[1, 1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))

# Color the box plots
box_colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(boxes['boxes'], box_colors):
    patch.set_facecolor(color)


# Adjust layout and save figure
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle
plt.savefig('Superstore_Audit.png', dpi=300)

plt.show()
print("Dashboard generated and saved as 'Superstore_Audit.png'.")
```
