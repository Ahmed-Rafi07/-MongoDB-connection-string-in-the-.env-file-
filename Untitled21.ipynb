### Task 1: Setup and Data Loading

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_df = sns.load_dataset('titanic')

# Display the first 5 rows
display(titanic_df.head())
```

### Task 2: Frequency Distribution with a Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['fare'], bins=50, kde=False)
plt.title('Distribution of Titanic Passenger Fares')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Task 3: Smooth Density with a KDE Overlay

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['fare'], bins=50, kde=True)
plt.title('Fare Distribution with KDE Overlay')
plt.xlabel('Fare')
plt.ylabel('Density / Number of Passengers')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Task 4: Analyzing Spread and Outliers

```python
plt.figure(figsize=(15, 7))
plt.suptitle('Spread and Distribution of Passenger Fares', fontsize=16)

# Box Plot
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.boxplot(y=titanic_df['fare'])
plt.title('Box Plot of Fares')
plt.ylabel('Fare')

# Violin Plot
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.violinplot(y=titanic_df['fare'])
plt.title('Violin Plot of Fares')
plt.ylabel('Fare')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent suptitle overlap
plt.show()
```
