```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="dark", palette="viridis")
penguins = sns.load_dataset('penguins')

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

fig.suptitle(
    "Palmer Penguins: Statistical Overview",
    fontsize=24,
    fontweight='bold'
)

sns.countplot(
    data=penguins,
    x="island",
    hue="species",
    ax=axes[0, 0]
)

axes[0, 0].set_title("Population Distribution by Island")

sns.violinplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue="sex",
    split=True,
    inner="stick",
    ax=axes[0, 1]
)

axes[0, 1].set_title("Body Mass Variance by Gender")

sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    size="body_mass_g",
    sizes=(20, 200),
    alpha=0.7,
    ax=axes[1, 0]
)

axes[1, 0].set_title("Bill Morphometrics vs. Body Mass")

sns.kdeplot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    fill=True,
    ax=axes[1, 1]
)

axes[1, 1].set_title("Flipper Length Density")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```
