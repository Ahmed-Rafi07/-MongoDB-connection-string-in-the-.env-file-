## Generate Dataset

### Subtask:
Generate the initial dataset as provided in Step 0 of the problem description.


**Reasoning**:
The first instruction is to import the `pandas` library. This is a prerequisite for creating and manipulating DataFrames.



```python
import pandas as pd
print("Pandas library imported.")
```

**Reasoning**:
The next step is to create the DataFrame `df` with the specified columns, data types, and populate it with sample data as instructed in the subtask.



```python
data = {
    'Transaction_ID': [1001, 1002, 1003, 1004, 1005],
    'Customer_ID': [1, 2, 1, 3, 2],
    'Transaction_Amount': [50.75, 120.00, 30.50, 200.25, 75.00],
    'Transaction_Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Product_Category': ['Electronics', 'Home Goods', 'Electronics', 'Apparel', 'Home Goods'],
    'Customer_Type': ['New', 'Existing', 'Existing', 'New', 'Existing']
}

df = pd.DataFrame(data)

df['Transaction_ID'] = df['Transaction_ID'].astype(int)
df['Customer_ID'] = df['Customer_ID'].astype(int)
df['Transaction_Amount'] = df['Transaction_Amount'].astype(float)
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
df['Product_Category'] = df['Product_Category'].astype(str)
df['Customer_Type'] = df['Customer_Type'].astype(str)

print("DataFrame 'df' created with the following structure:")
print(df.head())
```

## Detect Outliers - Z-Score

### Subtask:
Calculate the Z-scores for 'Transaction_Amount', identify outliers where absolute Z-score is greater than 3, and record the number of outliers, maximum outlier amount, and whether any wholesale transactions were flagged.


**Reasoning**:
The first step in calculating Z-scores is to determine the mean of the 'Transaction_Amount' column, as requested in the instructions.



```python
mean_transaction_amount = df['Transaction_Amount'].mean()
print(f"Mean of Transaction_Amount: {mean_transaction_amount}")
```

**Reasoning**:
The next step is to calculate the standard deviation of the 'Transaction_Amount' column, which is essential for computing the Z-scores as per the subtask instructions.



```python
std_transaction_amount = df['Transaction_Amount'].std()
print(f"Standard Deviation of Transaction_Amount: {std_transaction_amount}")
```

**Reasoning**:
Now that the mean and standard deviation are available, the next logical step is to compute the Z-score for each 'Transaction_Amount' and store it in a new column named 'Z_Score_Amount', as instructed by the subtask.



```python
df['Z_Score_Amount'] = (df['Transaction_Amount'] - mean_transaction_amount) / std_transaction_amount
print("Z-scores for 'Transaction_Amount' calculated and added to the DataFrame:")
print(df[['Transaction_Amount', 'Z_Score_Amount']].head())
```

**Reasoning**:
The next step is to identify outliers by filtering the DataFrame for transactions where the absolute value of 'Z_Score_Amount' is greater than 3, as specified in the subtask instructions.



```python
outliers = df[abs(df['Z_Score_Amount']) > 3]
print("Identified outliers based on Z-score > 3:")
print(outliers)
```

**Reasoning**:
Since outliers have been identified (or not, in this case), the next logical step is to count them and find the maximum transaction amount among them, while also addressing the wholesale transaction flagging as per the instructions.



```python
num_outliers = len(outliers)

if num_outliers > 0:
    max_outlier_amount = outliers['Transaction_Amount'].max()
else:
    max_outlier_amount = None # Or 0, depending on desired representation for no outliers

wholesale_flagged_statement = "No wholesale transactions were flagged as outliers using this method with the current data, as the dataset does not contain a 'Wholesale' customer type."

print(f"Number of outliers: {num_outliers}")
print(f"Maximum outlier amount: {max_outlier_amount}")
print(wholesale_flagged_statement)
```

## Detect Outliers - IQR

### Subtask:
Calculate Q1, Q3, and IQR for 'Transaction_Amount', define lower and upper fences, identify outliers outside these fences, and record the number of outliers, maximum outlier amount, and whether any wholesale transactions were flagged.


**Reasoning**:
The first step is to calculate the first quartile (Q1), the third quartile (Q3), and the Interquartile Range (IQR) for the 'Transaction_Amount' column as per the instructions, which are fundamental for defining outlier fences.



```python
Q1_amount = df['Transaction_Amount'].quantile(0.25)
Q3_amount = df['Transaction_Amount'].quantile(0.75)
IQR_amount = Q3_amount - Q1_amount

print(f"Q1 (25th percentile) of Transaction_Amount: {Q1_amount}")
print(f"Q3 (75th percentile) of Transaction_Amount: {Q3_amount}")
print(f"IQR (Interquartile Range) of Transaction_Amount: {IQR_amount}")
```

**Reasoning**:
Now that Q1, Q3, and IQR are calculated, the next step is to define the lower and upper fences for outlier detection using the IQR method, and then identify the outliers based on these bounds as per the subtask instructions.



```python
lower_bound_iqr = Q1_amount - 1.5 * IQR_amount
upper_bound_iqr = Q3_amount + 1.5 * IQR_amount

outliers_iqr = df[(df['Transaction_Amount'] < lower_bound_iqr) | (df['Transaction_Amount'] > upper_bound_iqr)]

print(f"Lower bound for IQR outlier detection: {lower_bound_iqr}")
print(f"Upper bound for IQR outlier detection: {upper_bound_iqr}")
print("Identified outliers based on IQR:")
print(outliers_iqr)
```

**Reasoning**:
Now that IQR outliers have been identified (or not), the next step is to calculate their count, find the maximum outlier amount if any exist, and generate the statement regarding wholesale transactions, as specified in the subtask instructions.



```python
num_outliers_iqr = len(outliers_iqr)

if num_outliers_iqr > 0:
    max_outlier_amount_iqr = outliers_iqr['Transaction_Amount'].max()
else:
    max_outlier_amount_iqr = None

wholesale_flagged_statement_iqr = "No wholesale transactions were flagged as outliers using this method with the current data, as the dataset does not contain a 'Wholesale' customer type."

print(f"Number of IQR outliers: {num_outliers_iqr}")
print(f"Maximum IQR outlier amount: {max_outlier_amount_iqr}")
print(wholesale_flagged_statement_iqr)
```

## Generate Comparison Table

### Subtask:
Create and populate the 'comparison_table' DataFrame with the results obtained from the Z-Score and IQR outlier detection methods.


**Reasoning**:
The next step is to create the comparison table by first defining its structure and then populating it with the results from both the Z-Score and IQR outlier detection methods, using the previously calculated variables, and finally displaying it.



```python
comparison_table = pd.DataFrame(columns=['Detection Method', 'Number of Outliers', 'Maximum Outlier Amount', 'Wholesale Flagged Statement'])

# Populate with Z-Score results
comparison_table.loc[0] = ['Z-Score', num_outliers, max_outlier_amount, wholesale_flagged_statement]

# Populate with IQR results
comparison_table.loc[1] = ['IQR', num_outliers_iqr, max_outlier_amount_iqr, wholesale_flagged_statement_iqr]

print("Comparison Table of Outlier Detection Methods:")
print(comparison_table)
```

## Apply Sanitization Policy

### Subtask:
Create a copy of the DataFrame. Apply the sanitization policy: remove transactions above ₹2,50,000, cap Wholesale transactions at ₹80,000, and create a new binary column 'Is_Outlier_IQR' based on IQR detection from the previous step.


**Reasoning**:
The first instruction is to create a new DataFrame `df_sanitized` as a copy of the original `df` DataFrame to apply sanitization policies without altering the original data.



```python
df_sanitized = df.copy()
print("DataFrame 'df_sanitized' created as a copy of 'df'.")
print(df_sanitized.head())
```

**Reasoning**:
The next step is to remove transactions from `df_sanitized` where 'Transaction_Amount' is greater than 250000.0, as specified in the subtask instructions.



```python
initial_rows = len(df_sanitized)
df_sanitized = df_sanitized[df_sanitized['Transaction_Amount'] <= 250000.0]
rows_removed = initial_rows - len(df_sanitized)

print(f"Removed {rows_removed} transactions where 'Transaction_Amount' was greater than 250000.0.")
print("First few rows of the sanitized DataFrame after removing high transactions:")
print(df_sanitized.head())
```

**Reasoning**:
The next step is to cap 'Wholesale' customer transactions at 80000.0, if their 'Transaction_Amount' exceeds this value, as per the sanitization policy described in the subtask. Although there are no 'Wholesale' customer types in the current dataset, the policy must still be applied.



```python
initial_sum_wholesale = df_sanitized.loc[df_sanitized['Customer_Type'] == 'Wholesale', 'Transaction_Amount'].sum()
df_sanitized.loc[(df_sanitized['Customer_Type'] == 'Wholesale') & (df_sanitized['Transaction_Amount'] > 80000.0), 'Transaction_Amount'] = 80000.0
final_sum_wholesale = df_sanitized.loc[df_sanitized['Customer_Type'] == 'Wholesale', 'Transaction_Amount'].sum()

if initial_sum_wholesale != final_sum_wholesale:
    print(f"Transactions for 'Wholesale' customers have been capped. Original sum: {initial_sum_wholesale}, New sum: {final_sum_wholesale}")
else:
    print("No 'Wholesale' transactions were capped as none exceeded the 80000.0 limit or no 'Wholesale' customers exist.")

print("First few rows of the sanitized DataFrame after capping wholesale transactions:")
print(df_sanitized.head())
```

**Reasoning**:
The final step for this subtask is to create a new binary column 'Is_Outlier_IQR' in `df_sanitized`, marking transactions as 1 if they were identified as IQR outliers (from `outliers_iqr`) and 0 otherwise, as per the subtask instructions.



```python
df_sanitized['Is_Outlier_IQR'] = df_sanitized['Transaction_ID'].isin(outliers_iqr['Transaction_ID']).astype(int)

print("Sanitized DataFrame with 'Is_Outlier_IQR' column:")
print(df_sanitized.head())
```

Create and populate the 'summary_table' DataFrame, comparing the mean and maximum 'Transaction_Amount' before and after applying the sanitization policy.
