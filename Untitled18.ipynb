```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.info()
```

```python
sns.set_theme(style = "whitegrid", palette= "muted",  context="notebook")
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", kde=True)
plt.title("Distribution of Flipper Lengths by Species")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.boxplot(data=penguins, x="species", y="body_mass_g", ax=ax[0], palette="Pastel1")
ax[0].set_title("Body Mass: Boxplot (Quartiles)")

sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex", split= True, inner = "quart", ax=ax[1], palette="Pastel2")
ax[1].set_title("Body Mass: Violinplot (Density & Sex)")
plt.show()
```

```python
g = sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
    fill=True,
    alpha=0.6
)

g.fig.suptitle("Correlation of Bill Dimensions", y=1.02)

plt.show()
```

```python
sns.lmplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    markers=["o", "s", "D"],
    height=6
)

plt.title("Linear Regression: Body Mass vs Flipper Length")

plt.show()
```

```python
sns.pairplot(
    data=penguins,
    hue="species",
    corner=True,
    palette="husl"
)

plt.show()
```
