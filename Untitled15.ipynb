```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('tips')

# --- TASK 1 CODE ---
# Create 'Spending_Type' column
df['Spending_Type'] = df['total_bill'].apply(lambda x: 'Premium' if x > 20 else 'Budget')

# Calculate 'tip_pct'
df['tip_pct'] = (df['tip'] / df['total_bill']) * 100

# Display the first few rows with new columns to verify
print("DataFrame after Task 1:")
display(df.head())

# --- TASK 2 CODE ---
# Create a summary table (Pivot Table or Groupby)
summary_table = df.groupby(['day', 'Spending_Type']).agg(
    median_tip_percentage=('tip_pct', 'median'),
    average_party_size=('size', 'mean')
).reset_index()

print("\nSummary Table (Task 2):")
display(summary_table)

# --- TASK 3 CODE ---
# Calculate Global Average Tip % for the entire dataset
global_average_tip_pct = df['tip_pct'].mean()

# Calculate Average Tip % for 'Budget' and 'Premium' diners
budget_average_tip_pct = df[df['Spending_Type'] == 'Budget']['tip_pct'].mean()
premium_average_tip_pct = df[df['Spending_Type'] == 'Premium']['tip_pct'].mean()

# Calculate the Difference (Delta)
delta_tip_pct = premium_average_tip_pct - budget_average_tip_pct

print(f"\nGlobal Average Tip %: {global_average_tip_pct:.2f}%")
print(f"Average Tip % for Budget Diners: {budget_average_tip_pct:.2f}%")
print(f"Average Tip % for Premium Diners: {premium_average_tip_pct:.2f}%")
print(f"Difference (Delta) between Premium and Budget Diners' Average Tip %: {delta_tip_pct:.2f}%")
```
