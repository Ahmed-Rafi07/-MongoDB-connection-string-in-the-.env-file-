# Global Fitness App Revenue Evolution Analysis

## Starter Code: Dataset Generation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate dates for 36 months (Jan 2021 to Dec 2023)
dates = pd.date_range(start='2021-01-01', periods=36, freq='MS')

# Initial upward growth trend (first 24 months)
initial_growth_months = 24
initial_revenue = np.linspace(100000, 250000, initial_growth_months) + np.random.normal(0, 10000, initial_growth_months)

# Sharp structural break (January 2023 is month 25)
# Revenue drops significantly and operates at a lower baseline
post_break_months = 36 - initial_growth_months
post_break_baseline = initial_revenue[-1] * 0.6 # 40% drop from last pre-break month
post_break_revenue = np.linspace(post_break_baseline, post_break_baseline * 1.05, post_break_months) + np.random.normal(0, 5000, post_break_months)

# Combine revenues
monthly_revenue = np.concatenate((initial_revenue, post_break_revenue))

# Create DataFrame
df_revenue = pd.DataFrame({
    'Date': dates,
    'Monthly_Revenue': monthly_revenue.round(2)
})

print("df_revenue DataFrame generated:")
display(df_revenue.head())
display(df_revenue.tail())
```

## Task 1: Trend Phase Computation

```python
# Split the df_revenue DataFrame into two new DataFrames
df_pre_break = df_revenue[(df_revenue['Date'] >= '2021-01-01') & (df_revenue['Date'] <= '2022-12-31')].copy()
df_post_break = df_revenue[(df_revenue['Date'] >= '2023-01-01') & (df_revenue['Date'] <= '2023-12-31')].copy()

# Metric Computation for pre-break phase
pre_mean_revenue = df_pre_break['Monthly_Revenue'].mean()
pre_std_revenue = df_pre_break['Monthly_Revenue'].std()
pre_first_month_revenue = df_pre_break['Monthly_Revenue'].iloc[0]
pre_last_month_revenue = df_pre_break['Monthly_Revenue'].iloc[-1]
pre_growth_pct = ((pre_last_month_revenue - pre_first_month_revenue) / pre_first_month_revenue) * 100

# Metric Computation for post-break phase
post_mean_revenue = df_post_break['Monthly_Revenue'].mean()
post_std_revenue = df_post_break['Monthly_Revenue'].std()
post_first_month_revenue = df_post_break['Monthly_Revenue'].iloc[0]
post_last_month_revenue = df_post_break['Monthly_Revenue'].iloc[-1]
post_growth_pct = ((post_last_month_revenue - post_first_month_revenue) / post_first_month_revenue) * 100

# Print Output
print(f"Pre-Break Mean: {pre_mean_revenue:.2f}")
print(f"Pre-Break Std: {pre_std_revenue:.2f}")
print(f"Pre-Break Growth %: {pre_growth_pct:.2f}")
print("")
print(f"Post-Break Mean: {post_mean_revenue:.2f}")
print(f"Post-Break Std: {post_std_revenue:.2f}")
print(f"Post-Break Growth %: {post_growth_pct:.2f}")
```

## Task 2: Structural Break Impact

```python
# Revenue Extraction
revenue_dec_2022 = df_revenue[df_revenue['Date'] == '2022-12-01']['Monthly_Revenue'].iloc[0]
revenue_jan_2023 = df_revenue[df_revenue['Date'] == '2023-01-01']['Monthly_Revenue'].iloc[0]

# Impact Calculation
absolute_drop = revenue_dec_2022 - revenue_jan_2023
percentage_drop = (absolute_drop / revenue_dec_2022) * 100

# Print Output
print(f"Revenue (Dec 2022): {revenue_dec_2022:.2f}")
print(f"Revenue (Jan 2023): {revenue_jan_2023:.2f}")
print(f"Absolute Revenue Drop: {absolute_drop:.2f}")
print(f"Percentage Revenue Drop: {percentage_drop:.2f}")
```

## Task 3: Rolling Average & Visualization

```python
# Create a new column for 3-month rolling average
df_revenue['Revenue_Rolling_3M'] = df_revenue['Monthly_Revenue'].rolling(window=3, min_periods=1).mean()

# Plotting
plt.figure(figsize=(14, 7))

# Plot Raw Monthly_Revenue
plt.plot(df_revenue['Date'], df_revenue['Monthly_Revenue'], label='Raw Monthly Revenue', color='blue', alpha=0.7)

# Plot 3-month rolling average
plt.plot(df_revenue['Date'], df_revenue['Revenue_Rolling_3M'], label='3-Month Rolling Average', color='orange', linewidth=2)

# Mark the structural break (January 2023)
plt.axvline(pd.to_datetime('2023-01-01'), color='red', linestyle='--', label='Structural Break (Jan 2023)')

# Add plot elements
plt.title('Global Fitness App Revenue: Raw vs. 3-Month Rolling Average', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Monthly Revenue ($)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
