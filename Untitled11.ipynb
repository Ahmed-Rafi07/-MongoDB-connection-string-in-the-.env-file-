```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 200

df = pd.DataFrame({
    "Ride_ID": [f"RIDE_{i+1}" for i in range(n)],
    "Ride_Distance": np.random.normal(8, 2.5, n).round(2),
    "Fare_Amount": np.random.normal(250, 80, n).round(2),
    "Driver_Rating": np.random.choice([3, 4, 5, np.nan], size=n, p=[0.2, 0.2, 0.12, 0.48]),
    "Vehicle_Type": np.random.choice(["Sedan", "SUV", "Mini", np.nan], size=n, p=[0.3, 0.3, 0.34, 0.06]),
    "Outside_Temperature": np.random.normal(30, 4, n).round(1),
    "Ride_Time": pd.date_range(start="2023-01-01", periods=n, freq="H")
})

# Introduce additional missingness
df.loc[df.sample(frac=0.03).index, "Ride_Distance"] = np.nan
df.loc[df.sample(frac=0.01).index, "Fare_Amount"] = np.nan
df.loc[df.sample(frac=0.12).index, "Outside_Temperature"] = np.nan

df.head()
```

```python
missing_percentage = (df.isnull().sum()/len(df)*100)
missing_percentage
```

```python
strategy_table = pd.DataFrame({
    "Column":df.columns,
    "Data_type":df.dtypes.values,
    "Percent_Missing": missing_percentage.values,
    "Chosen_Strategy":[
        "None",
        "Median",
        "Mean",
        "Flag_and_Zero",
        "Mode",
        "Forward_Fill",
        "None"
    ]
})
strategy_table
```

```python
original_df = df.copy()
```

```python
df["Ride_Distance"].fillna(df["Ride_Distance"].median(), inplace=True)
```

```python
df["Fare_Amount"].fillna(df["Fare_Amount"].mean(), inplace=True)
```

```python
df["Vehicle_Type"] = df["Vehicle_Type"].fillna(df["Vehicle_Type"].mode()[0])
```

```python
df.sort_values("Ride_Time")
```

```python
df["Outside_Temperature"].fillna(method="ffill", inplace=True)
```

```python
df["Driver_Rating"].fillna(0, inplace=True)
```

## Calculate Average Fare for Missing Ratings

### Subtask:
Calculate the average 'Fare_Amount' for rides where the 'Driver_Rating' was originally missing (using the 'Driver_Rating_Missing' indicator column).


**Reasoning**:
First, I need to identify the rows in the `original_df` where 'Driver_Rating' was missing. Then, I will filter the DataFrame using this condition and calculate the mean of the 'Fare_Amount' for these filtered rows, storing the result in a variable.



```python
avg_fare_missing_rating = original_df[original_df['Driver_Rating'].isnull()]['Fare_Amount'].mean()
print(f"Average Fare for rides with originally missing Driver Rating: {avg_fare_missing_rating:.2f}")
```

**Reasoning**:
Now that the average fare for missing driver ratings has been calculated, I will calculate the average 'Fare_Amount' for rides where the 'Driver_Rating' was originally provided (not null) in the `original_df`.



```python
avg_fare_provided_rating = original_df[original_df['Driver_Rating'].notnull()]['Fare_Amount'].mean()
print(f"Average Fare for rides with originally provided Driver Rating: {avg_fare_provided_rating:.2f}")
```

**Reasoning**:
I need to calculate the count of rides for both categories: where 'Driver_Rating' was originally missing and where it was originally provided. This will involve filtering the `original_df` and using `len()` or `.shape[0]`.



```python
count_missing_rating = len(original_df[original_df['Driver_Rating'].isnull()])
count_provided_rating = len(original_df[original_df['Driver_Rating'].notnull()])

print(f"Number of rides with originally missing Driver Rating: {count_missing_rating}")
print(f"Number of rides with originally provided Driver Rating: {count_provided_rating}")
```

**Reasoning**:
Now that all the required values (average fares and counts for both categories) have been calculated, I will create a summary table (`analysis_table`) to present these results in a structured format using a pandas DataFrame.



```python
analysis_table = pd.DataFrame({
    'Driver_Rating_Status': ['Originally Missing', 'Originally Provided'],
    'Average_Fare_Amount': [avg_fare_missing_rating, avg_fare_provided_rating],
    'Ride_Count': [count_missing_rating, count_provided_rating]
})

analysis_table
```
